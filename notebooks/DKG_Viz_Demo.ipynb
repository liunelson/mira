```python
import requests

from IPython.display import Image

from mira.dkg.viz import draw_relations
```

## Simple DKG relations visualization
Results returned from a simple relations query are visualized. Here, only the CURIEs (identifiers) are available.

```python
data_json ={
    "source_curie":"probonto:k0000016"
}
response = requests.post("http://34.230.33.149:8771/api/relations", json=data_json)
response
```

```python
draw_relations(response.json(), 'graph.png')

Image('graph.png')
```

## More Informative Node and Edge labels
If we add the key "full" to the payload, we can display the graph with named nodes and edges.

```python
data_json ={
    "source_curie":"probonto:k0000016",
    "full": True
}
response = requests.post("http://34.230.33.149:8771/api/relations", json=data_json)
response
```

```python
draw_relations(response.json(), 'graph.png', is_full=True)

Image('graph.png')
```
